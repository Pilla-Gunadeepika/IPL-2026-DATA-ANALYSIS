# INTRODUCTION
This project analyzes IPL 2026 match data using Python, Pandas, NumPy, and Matplotlib. The objective of the project is to explore team performances, batting statistics, bowling statistics, and venue-based insights through data analysis and visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

```python
matches.shape
```

```python
deliveries.shape
```

The matches dataset contains 39 matches and 23 columns. The deliveries dataset contains 17,477 ball-by-ball records and 19 columns. These datasets were explored before performing the analysis.

# **Analysis 1: Match Winners Analysis**

```python
matches['match_winner'].value_counts()
```

```python
matches['match_winner'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Matches Won By Each Team')
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.xticks(rotation=45)

plt.show()
```

# **OBSERVATION**

*   RCB and PBKS have won highest number of matches (6 each)
*   MI and KKR have won least number of matches (2 each)



# **Analysis 2: Top 10 Run Scorers**

```python
top_runs = deliveries.groupby('striker')['runs_of_bat'].sum()

top_runs.sort_values(ascending=False).head(10)
```

```python
top_runs.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Run Scorers')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**
Vaibhav Sooryavanshi emerged as the leading run scorer with 776 runs, followed by Shubman Gill (732) and Sai Sudharsan (722). The top-order batters dominated the season, with all top 10 players scoring more than 500 runs.

# **Analysis 3: Players with Most Sixes**

```python
sixes = deliveries[deliveries['runs_of_bat'] == 6]
most_sixes = sixes['striker'].value_counts()
most_sixes.sort_values(ascending=False).head(10)
```

```python
most_sixes.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Top 10 Players with most number of sixes')
plt.xlabel('Players')
plt.ylabel('No of sixes')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**


*   Vaibhav Sooryavanshi is clearly the leading six-hitter with around 72 sixes.
*   There is a significant gap between the first-ranked player and the rest.
*   Several players are clustered in the 30–45 sixes range, showing strong power-hitting performances.
*   The top 10 batters contributed heavily through boundary scoring.





# **Analysis 4: Players with Most Fours**

```python
fours = deliveries[deliveries['runs_of_bat'] == 4]
most_fours = fours['striker'].value_counts()
most_fours.sort_values(ascending=False).head(10)
```

```python
most_fours.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Top 10 Players with most number of fours')
plt.xlabel('Players')
plt.ylabel('No of fours')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**


Sai Sudharsan leads the tournament in fours, followed closely by Shubman Gill. This indicates their ability to score consistently through boundaries.



# **Analysis 5: Top Wicket Takers**

```python
wickets=deliveries[deliveries['player_dismissed'].notna()]
most_wickets=wickets['bowler'].value_counts()
most_wickets.sort_values(ascending=False).head(10)
```

```python
most_wickets.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Top 10 Players with most number of wickets')
plt.xlabel('Players')
plt.ylabel('No of wickets')
plt.xticks(rotation=45)
plt.show()
```

#OBSERVATION
Bhuvneshwar Kumar, Kagiso Rabada, and Jofra Archer emerged as the joint highest wicket-takers with 29 wickets each. The top bowlers consistently provided breakthroughs, making a significant impact on their teams' performances.

# **Analysis 6: Team-wise Total Runs**

```python
team_runs=deliveries.groupby('batting_team')['runs_of_bat'].sum()
team_runs=team_runs.sort_values(ascending=False)
team_runs
```

```python
team_runs.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title('Total runs made by team in tournament')
plt.xlabel('teams')
plt.ylabel('No of runs')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**
Rajasthan Royals (RR) scored the highest number of runs (2951) in the tournament, followed closely by Gujarat Titans (GT) with 2935 runs. RCB and SRH also displayed strong batting performances, while KKR recorded the lowest total runs among all teams.

# **Analysis 7: Venue-wise Total Runs**

```python
venue_runs=deliveries.groupby('venue')['runs_of_bat'].sum()
venue_runs=venue_runs.sort_values(ascending=False)
venue_runs
```

```python
venue_runs.plot(
    kind='bar',
    figsize=(12,5)
)
plt.title('VENUE WISE RUNS')
plt.xlabel('venue')
plt.ylabel('No of runs')
plt.show()
```

## **OBSERVATION**
Mumbai produced the highest number of runs in the tournament, followed by Hyderabad and Ahmedabad. On the other hand, Guwahati and Raipur recorded the lowest run totals. This suggests that batting conditions varied considerably across venues, with some grounds being more favorable for high-scoring matches.

# **Analysis 8: Venue Hosting Most Matches**

```python
matches['venue'].value_counts()
```

```python
matches['venue'].value_counts().plot(
    kind='bar',
    figsize=(12,5)
)
plt.title('VENUE HOSTED THE MOST MATCHES')
plt.xlabel('venue')
plt.ylabel('No of times hosted a match')
plt.show()
```

# **OBSERVATION**
M. Chinnaswamy Stadium, Bengaluru hosted the highest number of matches in the tournament. This indicates that it was one of the most frequently used venues and played a crucial role in the successful conduct of the season.

# **Analysis 9: Team-wise Sixes**

```python
sixes['batting_team'].value_counts()
```

```python
sixes['batting_team'].value_counts().plot(
    kind='bar',
    figsize=(12,5)
)
plt.title('TEAM WISE SIXES')
plt.xlabel('team')
plt.ylabel('No of sixes')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**


*   Rajasthan Royals (RR) hit the highest number of sixes in the tournament, followed by Sunrisers Hyderabad (SRH) and Royal Challengers Bengaluru (RCB). This indicates that these teams adopted an aggressive batting strategy and relied heavily on power hitting to score runs quickly.
*   RR not only scored the highest total runs in the tournament but also hit the most sixes, highlighting the team's strong and aggressive batting performance throughout the season.



# **Analysis 10: Team-wise Wickets**

```python
wickets['bowling_team'].value_counts()
```

```python
wickets['bowling_team'].value_counts().plot(
    kind='bar',
    figsize=(12,5)
)
plt.title('TEAM WISE WICKETS')
plt.xlabel('team')
plt.ylabel('No of wickets')
plt.xticks(rotation=45)
plt.show()
```

# **OBSERVATION**
Gujarat Titans (GT) recorded the highest number of wickets in the tournament, followed by Royal Challengers Bengaluru (RCB) and Rajasthan Royals (RR). This indicates that these teams had strong bowling attacks capable of taking regular wickets and putting pressure on opposition batters.
